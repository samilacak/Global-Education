## Analysis and Visualisation 

### Corelation 

```python
# Sadece sayısal sütunları seç
numeric_columns = df.select_dtypes(include='number')

# Korelasyon matrisi
correlation_matrix = numeric_columns.corr()

# Üst üçgeni maskeliyoruz
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Korelasyon matrisini ısı haritası olarak görselleştirme
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5, mask=mask)
plt.title("Sayısal Sütunlar Arasındaki Korelasyon Matrisi (Üst Üçgen)")
plt.show()
```

##### OOSR_Pre0Primary_Age_Male and OOSR_Pre0Primary_Age_Female

### Birth Rate Analys

```python
numeric_columns = df.select_dtypes(include='number')

corelation_birthdate = numeric_columns.corr()['Birth_Rate']

top_positif_correlations_birthrate = corelation_birthdate[corelation_birthdate.index != 'Birth_Rate'].sort_values(ascending=False).head(5)

print("5 Columns with the Highest Correlation:\n")
print(top_positif_correlations_birthrate)

print("\n It shows that out-of-school rates have an impact on birth rates, but these effects are weak. \n It is important to remember that other factors can also affect these relationships.")
```

```python
top_negatif_correlations_birthrate = corelation_birthdate[corelation_birthdate.index != 'Birth_Rate'].sort_values().head(5)

print("5 Columns with the Lowest Correlation:\n")
print(top_negatif_correlations_birthrate)

print("\n As students' reading proficiency increases, the birth rate decreases. \n In general, it has been observed that the birth rate increases in countries close to the Equator.")
```

### Unemploymend Rate 

Analysis: Among the columns with the highest positive correlation with the unemployment rate are the "Completion Rate" columns. This may indicate that there may be a relationship between education completion rates and the unemployment rate in a country. So, we can say that as education completion rates increase, the unemployment rate also tends to increase. However, this only represents a statistical relationship and further in-depth analysis may be necessary to determine true cause-effect relationships.

```python
numeric_columns = df.select_dtypes(include='number')

# Get correlations with column "Unemployment_Rate"
correlations_positif = numeric_columns.corr()['Unemployment_Rate']

# Sort positive correlations
top_positive_correlations = correlations_positif[correlations_positif.index != 'Unemployment_Rate'].sort_values(ascending=False).head(5)

print("5 Columns with the Highest Positive Correlation:\n")
print(top_positive_correlations)
```

```python
correlations_negatif = numeric_columns.corr()['Unemployment_Rate']

# Sort negatif corelations 
top_negative_correlations = correlations_negatif[correlations_negatif.index != 'Unemployment_Rate'].sort_values().head(5)

print("The 5 Columns with the Least Correlation: \n ")
print(top_negative_correlations)

print(" \n It shows that there is a weak negative relationship between the longitude coordinates of countries and the unemployment rate. \n So, we can say that changes in east-west location are associated with a decrease in the unemployment rate. \n These correlations are statistically weak and require further analysis and contextual information to determine whether the relationships are causal.")
```

```python
import plotly.express as px
def plot_unemployment_rates(df):
    try:
        fig = px.choropleth(df, 
                            locations='Countries and areas', 
                            locationmode='country names',
                            color='Unemployment_Rate',  # Choose the column for unemployment rates
                            title='Unemployment Rate by Country',
                            color_continuous_scale='purples',  # Use a purple color scale
                            range_color=[df['Unemployment_Rate'].min(), 
                                         df['Unemployment_Rate'].max()],  # Set the range of values
                            labels={'Unemployment_Rate': 'Unemployment Rate'}
                            )

        fig.update_geos(showocean=True, oceancolor="LightBlue", showland=True, landcolor="LightGreen")
        fig.show()

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
plot_unemployment_rates(df)
```
